```python
import matplotlib.pyplot as plt  # Импортируем библиотеку для построения графиков
plt.style.use('classic')  # Устанавливаем стиль графиков "classic"
%matplotlib inline  # Показываем графики прямо в Jupyter Notebook
import numpy as np  # Импортируем библиотеку NumPy для работы с массивами
import pandas as pd  # Импортируем библиотеку Pandas для работы с таблицами
```

```python
# Create some data
rng = np.random.RandomState(0)  # Создаем генератор случайных чисел с фиксированным seed (для воспроизводимости)
x = np.linspace(0, 10, 500)  # Генерируем 500 чисел от 0 до 10 с равными интервалами
y = np.cumsum(rng.randn(500, 6), 0)  # Создаем массив 500x6 случайных чисел и вычисляем их кумулятивную сумму по строкам
```

```python
# Plot the data with Matplotlib defaults
plt.plot(x, y)  # Строим графики для каждого из 6 столбцов массива y по оси x
plt.legend('ABCDEF', ncol=2, loc='upper left')  # Добавляем легенду с метками 'A'–'F', расположенную в верхнем левом углу, в 2 столбца
```

```python
import seaborn as sns  # Импортируем библиотеку Seaborn для улучшенной визуализации данных
sns.set_style('darkgrid')  # Устанавливаем стиль с темной сеткой
```

```python
# same plotting code as above!
plt.plot(x, y)  # Строим графики для каждого из 6 столбцов массива y по оси x
plt.legend('ABCDEF', ncol=2, loc='upper left')  # Добавляем легенду с метками 'A'–'F', расположенную в верхнем левом углу, в 2 столбца
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)  # Генерируем 2000 точек из двумерного нормального распределения
data = pd.DataFrame(data, columns=['x', 'y'])  # Создаем DataFrame с колонками 'x' и 'y'

for col in 'xy':  # Перебираем колонки 'x' и 'y'
    plt.hist(data[col], density=True, alpha=0.5)  # Строим нормированную гистограмму, делаем ее полупрозрачной
```

```python
for col in 'xy':  # Перебираем колонки 'x' и 'y'
    sns.kdeplot(data[col], shade=True, color='red')  # Строим график оценки плотности распределения (KDE) с заливкой
```

```python
sns.distplot(data['x'])  # Строим гистограмму с оценкой плотности для данных 'x'
sns.distplot(data['y'])  # Строим гистограмму с оценкой плотности для данных 'y'
```

```python
sns.kdeplot(data);  # Строим график оценки плотности распределения (KDE) для данных в DataFrame 'data' (по умолчанию для всех колонок)
```

```python
import datetime

def convert_time(s):
    h, m, s = map(int, s.split(':'))
    return datetime.timedelta(hours=h, minutes=m, seconds=s)

data = pd.read_csv('marathon-data.csv',
                   converters={'split':convert_time, 'final':convert_time})
data.head()
```

```python
data.dtypes
```

```python
data['split_sec'] = data['split'].astype(int) / 1E9
data['final_sec'] = data['final'].astype(int) / 1E9
data.head()
```

```python
with sns.axes_style('white'):
    g = sns.jointplot("split_sec", "final_sec", data, kind='hex')
    g.ax_joint.plot(np.linspace(4000, 16000),
                    np.linspace(8000, 32000), ':k')
```

```python
data['split_frac'] = 1 - 2 * data['split_sec'] / data['final_sec']
data.head()
```

```python
sns.distplot(data['split_frac'], kde=False);
plt.axvline(0, color="k", linestyle="--");
```

```python
sum(data.split_frac < 0)
```

```python
g = sns.PairGrid(data, vars=['age', 'split_sec', 'final_sec', 'split_frac'],
                 hue='gender', palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```

```python
sns.kdeplot(data.split_frac[data.gender=='M'], label='men', shade=True)
sns.kdeplot(data.split_frac[data.gender=='W'], label='women', shade=True)
plt.xlabel('split_frac');
```

```python
sns.violinplot("gender", "split_frac", data=data,
               palette=["lightblue", "lightpink"]);
```

```python
data['age_dec'] = data.age.map(lambda age: 10 * (age // 10))
data.head()
```

```python
men = (data.gender == 'M')
women = (data.gender == 'W')

with sns.axes_style(style=None):
    sns.violinplot("age_dec", "split_frac", hue="gender", data=data,
                   split=True, inner="quartile",
                   palette=["lightblue", "lightpink"]);
```
