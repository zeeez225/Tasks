Yahoo Finance предоставляет API для доступа к финансовым данным, которые могут быть использованы для анализа в Python. Для начала работы необходимо зарегистрироваться на сайте Yahoo Finance и получить свой API-ключ.

Для получения данных о котировках определенной компании за определенный период времени мы можем использовать следующий код:

```python
import yfinance as yf

# Задаем интервал времени, за который хотим получить данные
start_date = '2022-01-01'
end_date = '2022-03-31'

# Получаем данные о котировках TSLA за указанный период
ticker = 'TSLA'
data = yf.download(ticker, start=start_date, end=end_date)

# Выводим первые 5 строк полученных данных
print(data.head())
```

Документация Yahoo Finance API содержит полный список методов и параметров, которые можно использовать для получения данных о финансовом рынке. Некоторые из основных методов API включают:

Ticker(symbol) - создает объект, который содержит данные о компании, связанной с указанным символом. Например, Ticker("TSLA") создаст объект, содержащий данные об Tesla, Inc..

history(start=None, end=None, interval="1d", actions=True, auto_adjust=True) - возвращает исторические данные о котировках акций для указанного периода. Метод принимает параметры start и end для задания начальной и конечной даты периода, параметр interval для задания интервала (1d, 1wk, 1mo), параметр actions для включения/отключения данных о дивидендах и слияниях компаний, параметр auto_adjust для автоматической корректировки цен при слияниях и дивидендах.

info - возвращает информацию о компании, связанной с текущим объектом Ticker.

financials - возвращает финансовые показатели компании.

recommendations - возвращает список рекомендаций аналитиков по акциям компании.

calendar - возвращает календарь событий, связанных с компанией, таких как отчеты о прибыли, конференции для инвесторов и т.д.

news - возвращает список новостей, связанных с компанией.

options - возвращает информацию о опционах на акции компании.

Для работы с API необходимо использовать ключ API, который можно получить после регистрации на сайте Yahoo Finance. Ключ API можно использовать для доступа к различным методам API и получения данных в формате JSON или CSV.

Некоторые из ограничений использования Yahoo Finance API включают:

Ограниченная пропускная способность. Yahoo Finance API может ограничивать количество запросов, которые можно отправлять за определенный период времени.

Ограничения на коммерческое использование. Yahoo Finance API не может использоваться для коммерческих целей без предварительного разрешения.

Ограничения на использование в продуктах, конкурирующих с продуктами Yahoo Finance. Yahoo Finance API не может использоваться в продуктах, которые конкурируют с продуктами Yahoo Finance без предварительного разрешения.

Для получения данных о котировках необходимо отправить запрос на API с указанием нужных параметров. Например, чтобы получить данные о ценах закрытия акций Apple за последние 30 дней, можно использовать следующий код на Python:

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Создание объектов Ticker для заданных компаний
tsla = yf.Ticker("TSLA")
aapl = yf.Ticker("AAPL")

# Получение цены закрытия акций за последние 5 лет
tsla_hist = tsla.history(period="5y")
aapl_hist = aapl.history(period="5y")

# Объединение таблиц цен закрытия акций для обеих компаний
merged = pd.merge(tsla_hist["Close"], aapl_hist["Close"], left_index=True, right_index=True, suffixes=("_TSLA", "_AAPL"))

# Вывод графиков цен закрытия акций обеих компаний
merged.plot()
plt.title("TSLA vs AAPL Close Prices")
plt.show()

# Рассчет коэффициента корреляции между ценами закрытия акций обеих компаний
corr = merged.corr().iloc[0,1]

# Вывод коэффициента корреляции
print("TSLA vs AAPL Correlation: {:.2f}".format(corr))
```

Анализ изменения цены акций нескольких компаний в заданный период времени и построение графика совместных доходностей

```python

import yfinance as yf
import matplotlib.pyplot as plt
import datetime as dt
import pandas as pd

# Создание списка из нескольких компаний
tickers = ["MSFT", "AAPL", "AMZN", "TSLA","FRHC"]

# Получение цены закрытия акций за заданный период времени для каждой компании
start_date = "2021-01-01"
end_date = "2021-12-31"
closing_prices = {}
for ticker in tickers:
    stock = yf.Ticker(ticker)
    hist = stock.history(start=start_date, end=end_date)
    closing_prices[ticker] = hist["Close"]

# Создание DataFrame из цен закрытия акций
df = pd.DataFrame(closing_prices)

# Рассчет дневных доходностей для каждой компании
daily_returns = df.pct_change()

# Рассчет совместных доходностей
weights = [0.25, 0.25, 0.25, 0.25, 0.25] # равные веса для каждой компании
portfolio_returns = (daily_returns * weights).sum(axis=1)

# Построение графика совместных доходностей
plt.plot(portfolio_returns)
plt.title("Portfolio Returns for 2021")
plt.xlabel("Date")
plt.ylabel("Returns")
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Задание тикеров двух компаний
ticker1 = "TSLA"
ticker2 = "AAPL"

# Получение цен закрытия за заданный период времени для двух компаний
start_date = "2021-01-01"
end_date = "2021-12-31"
stock1 = yf.Ticker(ticker1)
stock2 = yf.Ticker(ticker2)
stock1_hist = stock1.history(start=start_date, end=end_date)
stock2_hist = stock2.history(start=start_date, end=end_date)

# Создание DataFrame из цен закрытия для двух компаний
df = pd.concat([stock1_hist["Close"], stock2_hist["Close"]], axis=1)
df.columns = [ticker1, ticker2]

# Рассчет дневных доходностей для двух компаний
df_returns = df.pct_change()

# Рассчет корреляции между доходностями двух компаний
corr = df_returns.corr().iloc[0,1]

# Построение scatter plot
plt.scatter(df_returns[ticker1], df_returns[ticker2])
plt.title(f"{ticker1} vs {ticker2} - Correlation: {corr:.2f}")
plt.xlabel(f"{ticker1} Returns")
plt.ylabel(f"{ticker2} Returns")
plt.show()
```
